```python
import pandas as pd
import statsmodels.api as sm
df=pd.read_csv("Advertising.csv")
```

```python
df
```

```python
X=df[["TV","radio","newspaper"]]
y=df['sales']
```

```python
df.head()
```

```python
X
```

```python
y
```

```python
X=sm.add_constant(X)
```

```python
X
```

```python
# fit the OLS model with intercept on tV and radio
model=sm.OLS(y,X).fit()
```

```python
model.summary()
```
