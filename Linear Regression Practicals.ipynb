```python
## Independent And Dependent Features
X=dataset.iloc[:,:-1]
y=dataset.iloc[:,-1]
```

```python
X.head()
```

```python
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.33, random_state=10)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
## Standardize or feature scaling the datasets
from sklearn.preprocessing import StandardScaler
scaler=StandardScaler()
```

```python
scaler
```

```python
X_train=scaler.fit_transform(X_train)
```

```python
X_test=scaler.transform(X_test)
```

```python
X_train
```

```python
X_test
```

## Model Training

```python
from sklearn.linear_model import LinearRegression
```

```python
regression=LinearRegression()
```

```python
regression
```

```python
regression.fit(X_train,y_train)
```

```python
## print the coefficients and the intercept
print(regression.coef_)
```

```python
print(regression.intercept_)
```

```python
## PRediction for the test data
reg_pred=regression.predict(X_test)
```

```python
reg_pred
```

### Assumptions Of Linear Regression

```python
## residuals
residuals=y_test-reg_pred
residuals
```

```python
plt.scatter(y_test,reg_pred,color='blue', s=50, alpha=0.6, edgecolors='k', linestyle='--', linewidth=2, label='Perfect Prediction')
plt.xlabel("Actual Test Data")
plt.ylabel("Test Predicted Data")
plt.title("Actual vs Predicted Test Data", fontsize=16)
plt.legend(loc='upper left', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)
plt.xticks(fontsize=12)
plt.yticks(fontsize=12)
```

```python
residuals
```

```python
sns.displot(residuals,kind="kde")
```

```python
## SCatter plot with predictions and residual
##uniform distribution
plt.scatter(reg_pred,residuals)
```

```python
## Performance Metrics
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(mean_squared_error(y_test,reg_pred))
print(mean_absolute_error(y_test,reg_pred))
print(np.sqrt(mean_squared_error(y_test,reg_pred)))
```

## R square and adjusted R square

```python
from sklearn.metrics import r2_score
score=r2_score(y_test,reg_pred)
print(score)
```

```python
## Adjusted R square
#display adjusted R-squared
1 - (1-score)*(len(y_test)-1)/(len(y_test)-X_test.shape[1]-1)
```
